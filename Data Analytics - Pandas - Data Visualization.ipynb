```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Data Analysis\Pandas file\Data Visualization\File.csv")
df.set_index("Date")
```

```python
df
```

# Default Chart

```python
#plot mtd is used to convert our data to chart. By default, line chart will display.

df.plot()
```

# Line chart

```python
#Here we are mentioning the type of chart(kind attribute), title, xlabel and ylabel for the chart.

df.plot(kind="line",title="My Chart", xlabel="Date", ylabel="Ratings")
```

```python
#You can also use subplots, where each cols will be displayed in seperate charts

df.plot(kind="line", subplots=True)
```

# Bar chart

```python
#bar chart. If we put stacked as true, we would get stacked bar graph(with out that, it would be a vertical bar graph)

df.plot(kind="bar", stacked = True)
```

```python
#To get a chart for only one cols or specific cols. Here if there is one col, label in chart will not appear.

df["Flavor Rating"].plot(kind="bar")
```

```python
#For horizontal bar chart, we cannot put kind as True as we use mtd barh() to obatin it. We can use other attributes like stacked

df.plot.barh(stacked= True)
```

# Scatter plot

```python
#Scatter plot is plotted with respect to x and y axis, so its complusary for us to mention x and y axis to our cols present in df
#with out it, we will get error. Other parameters, s->size of our dots and c->color of dots

df.plot.scatter(x="Flavor Rating", y="Texture Rating", s=100, c="black")
```

# Histogram

```python
#Histogram tells us the distrubution of our data in set of numbers. By default, it will provide 10 bins. We can also change
#the same to other number of bins as well

df.plot.hist(bins=20)
```

# Box plot

```python
#Here bottom line is min value and top line is max value.
#bottom of box is 25% of value, middle line is 50% and top box line is 75%

df.boxplot()
```

# Area chart

```python
#Area chart is similar to line chart.
#figsize is an attribute used to increase overall size of chart. this can be used for all chart types

df.plot.area(figsize=(10,5))
```

# Pie chart

```python
#In pie chart, it is mandatory to provide y axis col, or it will throw error
df.plot.pie(y="Flavor Rating", figsize=(10,7))
```
